## 2020 ~ 2024년의 ESG 데이터 (한국ESG기준원)

    - 칼럼명( 기업명 / 기업코드 / ESG구조 / 환경 / 사회 / 지배구조 )

```python
import pandas as pd
import numpy as np
```

```python
# esg data 불러오기
df = pd.read_excel('esg_data.xlsx')
df1 = df.copy()
```

```python
# NO, ESG 등급조정, 환경 등급조정, 사회 등급조정, 지배구조 등급 조정 제거
df1.drop(
    columns={
        'NO','ESG 등급조정','환경 등급조정','사회 등급조정','지배구조 등급조정'
    }
    , inplace=True
)
```

```python
df1.info()
```

```python
flag = df1[df1.isnull().all(axis=1)].index
```

```python
flag
# [649, 775, 1921, 2397, 2667] -> 2667은 행 삭제 
```

```python
# 검색
# 649 : SK이노베이션 , 2021년도로 수정 
# df1.loc[648:650,]
df1.loc[649,'기업명'] = 'SK이노베이션'
df1.loc[649,'기업코드'] = '96770.0'
df1.loc[649,'평가년도'] = 2021

# 775 : 경동나비엔, 2021년도로 수정
# df1.loc[774:776,]
df1.loc[775,'기업명'] = '경동나비엔'
df1.loc[775,'기업코드'] = '9450.0'
df1.loc[775,'평가년도'] = 2021

# 1921 : 백광산업, 2021년도로 수정
# df1.loc[1920:1923,]
df1.loc[1921,'기업명'] = '백광산업'
df1.loc[1921,'기업코드'] = '1340.0'
df1.loc[1921,'평가년도'] = 2021

# 2397 : 세아베스틸지주로 이름 변경(2396도) 
# df1.loc[2396:2399,]
df1.loc[2396:2398,'기업명'] = '세아베스틸지주'
df1.loc[2397,'기업코드'] = '1430.0'
df1.loc[2397,'평가년도'] = 2021

# 2667 행 삭제 -> 어떤 회사인지 모름 
df1 = df1.drop(index=2667)

# 확인 
df1[df1.isnull().all(axis=1)]
```

```python
# 평가년도 int로 바꾸기
df1['평가년도'] = df1['평가년도'].astype('float').astype('int')
```

```python
# D등급 1점, C등급 2점, B등급 3점, B+등급 4점, A등급 5점, A+등급 6점
# 등급없음 NAN 처리 
# df1['ESG등급'].value_counts()
# df1['환경'].value_counts()
# df1['사회'].value_counts()
# df1['지배구조'].value_counts()
grade_map = {
    'D' : 1,
    'C' : 2,
    'B' : 3,
    'B+' : 4,
    'A' : 5,
    'A+' : 6,
    '등급없음' : 0
}

# ESG 등급 점수화
df1['ESG등급'] = df1['ESG등급'].map(grade_map)
df1['환경'] = df1['환경'].map(grade_map)
df1['사회'] = df1['사회'].map(grade_map)
df1['지배구조'] = df1['지배구조'].map(grade_map)
```

```python
df1.head()
```

```python
# 총 1338개의 unique한 기업명이 있다
list1 = df1['기업명'].unique().tolist()
# 리스트를 데이터프레임으로 변환
df_list = pd.DataFrame(list1,columns=['기업명'])
# CSV 파일로 저장
df_list.to_csv("기업명리스트.csv",index=False, encoding='utf-8-sig')
```

```python
# 금융업 제거 (118개 기업 제거) -> df1 5112개 행 -> df2 4627개 행(485개 제거)  
# 금융업(은행, 보험, 증권 등)은 일반 제조업이나 서비스업과는 
# 전혀 다른 회계 기준과 수익구조를 가지고 있습니다.
keywords = ['보험','은행','증권','지주','투자','캐피탈','NICE']
list2 = [name for name in df1['기업명'].unique() if any(keyword in name for keyword in keywords)]
print(list2)
# len(list2)
df2 = df1.copy()
for i in df2.index:
    if df2.loc[i,'기업명'] in list2:
        df2 = df2.drop(i)
```
