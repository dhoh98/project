```python
import pandas as pd
```

```python
# 한국 ESG 기준원 데이터 불러오기 
# 10990개 
evaluate_df = pd.read_excel('evaluate_data.xlsx',dtype={
    '기업코드' : str
})
evaluate_df.head()
```

```python
# ESG 데이터 컬럼확인
evaluate_df.info()
```

```python
# 기업명, 기업코드, ESG 등급, 환경, 사회, 지배구조, 평가년도 컬럼만 사용하자
ev_df = evaluate_df.drop(columns={
    'NO','ESG 등급조정','환경 등급조정','사회 등급조정','지배구조 등급조정'
})
ev_df.head()
```

```python
# 평가년도 확인 2012년은 제거 -> 706개 
# 2013년부터 ESG 등급체계 달라짐
# 10284개 행
ev_df = ev_df[~(ev_df['평가년도'] == 2012)]
# ev_df.to_excel('ev_df_v1.xlsx')
```

```python
# ESG등급 등급없음 제거 -> 345개 제거 : 9939개 
# ESG등급, 환경, 사회, 지배구조 모두 미공개면 제거 -> 18개 제거 : 9921개
ev_df = ev_df[~(ev_df['ESG등급'] == '등급없음')]
flag1 = (ev_df['ESG등급'] == '미공개')
flag2 = (ev_df['환경'] == '미공개')
flag3 = (ev_df['사회'] == '미공개')
flag4 = (ev_df['지배구조'] == '미공개')
ev_df[~flag1 & (flag2 | flag3 | flag4)]
```

```python
# 평가없음 = 0, D = 1, C = 2, B = 3, B+ = 4, A = 5, A+ = 6 점으로 계량화
for idx, row in ev_df.iterrows():
    print(idx)
    print(row)
```
